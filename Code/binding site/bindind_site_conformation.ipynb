```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
path = r"Side_map_2.xlsx"
import openpyxl
ref_workbook = openpyxl.load_workbook(path)
sheetnames = ref_workbook.sheetnames
print(sheetnames)
ws = ref_workbook[sheetnames[0]]
ws
```

```python
data = ws.values
# Get the first line in file as a header line
columns = next(data)[0:]
# Create a DataFrame based on the second and subsequent lines of data
df = pd.DataFrame(data, columns=columns)
#df['bindingsite'] = pd.to_numeric(df['bindingsite'])
df
```

```python
df.columns
```

```python
table = pd.pivot_table(df, values=['SiteScore', 'balance', 'Dscore', 'volume'], index=['bindingsite'], aggfunc=[np.mean, np.std], sort = False)
table.drop(labels = [12], axis = 0, inplace = True)
table= table.round(2)
table = table.sort_values(('mean', 'SiteScore'), ascending = False)
table
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer_siteScore_withSTD = pd.ExcelWriter('sorting_by_SiteScore_results_with_STD.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
table.to_excel(writer_siteScore_withSTD, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer_siteScore_withSTD.close()

print('I did it')
```

```python
bindingsite_grouping = df.groupby(['bindingsite'])['SiteScore','Dscore', 'volume', 'balance', 'size'].mean()
#deleting not nesessary binding site, which are too large 
bindingsite_grouping.drop(labels = [12], axis = 0, inplace = True)
bindingsite_grouping
```

```python
final_bindingsite_grouping= bindingsite_grouping.round(2)
sorting_by_SiteScore = final_bindingsite_grouping.sort_values('SiteScore', ascending = False)
sorting_by_SiteScore

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer_siteScore = pd.ExcelWriter('sorting_by_SiteScore_results.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
sorting_by_SiteScore.to_excel(writer_siteScore, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer_siteScore.close()

print('I did it')
```

```python
path = r"first_binding_site.xlsx"
import openpyxl
ref_workbook = openpyxl.load_workbook(path)
sheetnames = ref_workbook.sheetnames
print(sheetnames)
ws = ref_workbook[sheetnames[0]]
data = ws.values
# Get the first line in file as a header line
columns = next(data)[0:]
# Create a DataFrame based on the second and subsequent lines of data
df = pd.DataFrame(data, columns=columns)
df.head(10)
```

```python
analysis_df = pd.DataFrame(pd.to_numeric(df['SiteScore']))
analysis_df.mean()
```

```python
df['SiteScore'] = pd.to_numeric(df['SiteScore'])
df['balance'] = pd.to_numeric(df['balance'])
df_desc = df.describe()
df_desc_round = df_desc.round(2)
df_desc_round
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('first_binding_site_results.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df_desc_round.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.close()

print('I did it')
```
